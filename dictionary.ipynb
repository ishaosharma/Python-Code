```python
rec = {}
n = int(input("Enter number of total students : "))
i = 1
while i <= n:
    name = input("Enter students Name : ")
    marks = input("Enter percentage of marks : ")
    rec[name] = marks
    i = i + 1
print("Name of student: ", "\t", "% of marks: ")
for value in rec:
    print(value, "\t", rec[value])
```

```python
def outer():
    x = 777
    def inner():
        nonlocal x
        x = 700
        print("Value of x inside inner function is = ", x)
    inner()
    print("Value of x inside outer function = ", x)
        
outer()
```
