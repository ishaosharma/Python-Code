which numpy operation correct for square in array
1. arr * arr
2. arr ** 2
3. np.square(arr)
4. all of the above 
which option is correct?

The correct answer is:

**4. all of the above**

Explanation:
- `arr * arr` performs element-wise multiplication, effectively squaring each element.
- `arr ** 2` uses the power operator to square each element.
- `np.square(arr)` applies NumPy's square function element-wise to the array.

All three operations will produce the same result of squaring each element in the array.[1][3][5][7]



what does all() function do in numpy?

The np.all() function in NumPy tests whether all elements in an array evaluate to True. By default, it performs this check over the entire array, but you can specify an axis to check along a particular dimension. It returns a single Boolean value True if every element is True (or non-zero/non-false), otherwise False.â€‹

Key points:
np.all(arr) returns True only if every element in arr is True.

You can specify axis to apply the check along rows, columns, or specific axes.

Useful for conditions checking, filtering, or validating array contents.

```python
import numpy as np

arr = np.array([[True, True], [True, False]])
print(np.all(arr))          # Outputs: False
print(np.all(arr, axis=0))  # Outputs: [ True False]
print(np.all(arr, axis=1))  # Outputs: [ True False]
```

```python
import numpy as np

arr = np.array([[True, True], [True, False]])
print(np.all(arr))          # Outputs: False
print(np.all(arr, axis=0))  # Outputs: [ True False]
print(np.all(arr, axis=1))  # Outputs: [ True False]
```

Examples of np.all with axis parameter

Explanation:

axis=0 applies all() column-wise, reducing rows and returning a Boolean array for each column.

axis=1 applies all() row-wise, reducing columns and returning a Boolean array for each row.

Without axis, it checks all elements in the array at once

```python
import numpy as np

arr = np.array([[True, True, False],
                [True, True, True]])

# Check if all elements in the entire array are True
print(np.all(arr))  
# Output: False (because there is a False element)

# Check if all elements are True along axis 0 (columns)
print(np.all(arr, axis=0))  
# Output: [ True  True False] 
# Column-wise result: True only if all elements in that column are True

# Check if all elements are True along axis 1 (rows)
print(np.all(arr, axis=1))  
# Output: [False  True] 
# Row-wise: True only if all elements in that row are True
```
