```python
#Con 2: Function Aliasing
#EX 1
def fun1():
    print("Hello")
fun1()  #here fun1 pointing to fun object.
f2 = fun1 #Now f2 also pointing to same fun1 object.
print(id(fun1))
print(id(f2))
#Now we can call this function either by using
#f2 or by using fun1
f2()
```

```python
#Con 2: Function Aliasing
#EX 2
def fun1():
    print("Hello")
fun1()   #here fun1 pointing to fun object
f2 = fun1   #now f2 also pointing to same fun1 object
print(id(fun1))
print(id(f2))
#Now we can call this function either by using f2 or by using fun1
f2()
del f2      #deletes the reference that f2 holds and not the memory space
#Here i am deleting wish function
#But here we are only deleting reference variable
fun1()
f2()  #Now it is invalid because we deleted it.
```

```python
# Con3: Nested Function

def outer():
    print("This is outer fun")
    def inner():
        print("This is Inner")
    inner()     #calling inner fun
    return inner
k = outer()   #Calling outer fun    and k holds the returned inner function.
k()           #calling k so it executer inner function
# Then k() is called, which executes inner() again, printing "This is Inner" the second time.
# Therefore, "This is Inner" is printed twice â€” once inside outer() when inner() is called, 
# and the second time when k() calls inner() after outer() returns.



```
