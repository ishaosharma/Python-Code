```python
def searching_number_in_tuple(tuple):
    number = int(input())
    index = 0
    while(index < len(tuple)):
        if number == tuple[index]:
            print("Number found at index ", index)
        index += 1

searching_number_in_tuple((1,4,9,16,25,36,49,64,81,100))
```

WAP to remove duplicate characters from string 

```python
s = "ababababa"
final_string = ""
for char in s:
    if char not in final_string:
        final_string += char
print("Final string after removing duplicates = ", final_string)
    
```

```python
def remove_duplicates(string):
    final_string = ""
    for char in string:
        if char not in string:
            final_string += char
print(final_string)

remove_duplicates("ababababa")
```

```python
statement = [("credit", 600), ("debit", 200), ("credit", 600), ("debit", 200)]
total_credit_amount = 0
total_debit_amount = 0
for key, value in statement:
    if key == "credit":
        total_credit_amount += value
    else:
        total_debit_amount += value
print("Total credit amount is : ", total_credit_amount)
print("Total debit amount is : ", total_debit_amount)

```

```python
def statement_report(statements):
    total_credit = 0
    total_debit = 0
    for k, v in statements:
        if k == "credit":
            total_credit += v
        else:
            total_debit += v
    return total_credit, total_debit
    
statements = [("credit", 600), ("debit", 200), ("credit", 600), ("debit", 200)]
credit, debit = statement_report(statements)
print(credit)
print(debit)
```
