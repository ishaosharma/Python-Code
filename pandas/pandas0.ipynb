<h3>Create a simple Pandas Series from a list:</h3>

```python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)

print(myvar[0])
print(myvar[1])
```

```python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"]) 

print(myvar)

print(myvar["y"])
print(myvar["z"])
```
