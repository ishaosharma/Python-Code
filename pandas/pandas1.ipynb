Create a DataFrame using dict of dict

---

```python
import pandas as pd
names = ["Ajay","Amit","Ayushi","Aarvi","Puzzu"]
ages = [22,24,20,24,26] #create another list
fees = [True, False, True,True,False]
dict_names = {k:v for k,v in enumerate(names)}  #dictionary
print("dict_names : ",dict_names)
dict_ages = {k:v for k,v in enumerate(ages)}
print("dict_ages : ",dict_ages)
dict_fees = {k:v for k,v in enumerate(fees)}
print("dict_fees : ",dict_fees)

#create the pandas dataframe
students = pd.DataFrame({'Name':dict_names,'Age':dict_ages,'Fees Submitted': dict_fees})

#print dataframe
print(students)
```

what is zip()?

---

zip() is a built-in Python function that groups items from two or more sequences (like lists, tuples, or strings) into tuples, position by position. It returns an iterator of these tuples, which you usually convert to a list or use in a loop.Think of it like zipping two sides of a zipper together.

Simple definition
zip() takes multiple iterables and pairs the 1st elements together, then the 2nd elements, and so on.

It stops when the shortest iterable runs out.

Syntax: zip(iterable1, iterable2, ...)

Example 1: Combine two lists

```python
names = ["Amit", "John", "Jacob"]
marks = [95, 70, 80]

paired = list(zip(names, marks))
print(paired)
# Output: [('Amit', 95), ('John', 70), ('Jacob', 80)]
#Here, first elements are paired together, then second elements, and so on.
```

Example 2: Different lengths

```python
students = ["Amit", "John", "Jacob", "David"]
ranks = [1, 4, 3]

result = list(zip(students, ranks))
print(result)
# Output: [('Amit', 1), ('John', 4), ('Jacob', 3)]
#Extra items in the longer list are ignored because zip() stops at the shortest list.
```

Example 3: Using zip() in a loop

```python
students = ["Amit", "John", "Jacob"]
marks = [95, 70, 80]

for name, mark in zip(students, marks):
    print(name, "scored", mark)
#This is very useful for looping over multiple lists at the same time.
```

---

Example 4: Creating a dictionary with zip()

---

```python
students = ["Amit", "John", "Jacob"]
marks = [95, 70, 80]

marks_dict = dict(zip(students, marks))
print(marks_dict)
# Output: {'Amit': 95, 'John': 70, 'Jacob': 80}
#Here, zip() helps make a dictionary where one list becomes keys and the other becomes values.
```

zip() is a built-in Python function that groups items from two or more sequences (like lists, tuples, or strings) into tuples, position by position. It returns an iterator of these tuples, which you usually convert to a list or use in a loop.

Simple definition
zip() takes multiple iterables and pairs the 1st elements together, then the 2nd elements, and so on.

It stops when the shortest iterable runs out.

Syntax: zip(iterable1, iterable2, ...)

```python
names = ["Amit", "John", "Jacob"]
marks = [95, 70, 80]

result = list(zip(names, marks))
print(result)
# [('Amit', 95), ('John', 70), ('Jacob', 80)]
```

Using zip() in a loop
You can loop over multiple lists at the same time:

```python
students = ["Amit", "John", "Jacob"]
ranks = [1, 4, 3]

for name, rank in zip(students, ranks):
    print(name, "has rank", rank)
```

Creating a dictionary with zip()
One common use is to make a dictionary from two lists:

```python
students = ["Amit", "John", "Jacob"]
marks = [95, 70, 80]

student_marks = dict(zip(students, marks))
print(student_marks)
# {'Amit': 95, 'John': 70, 'Jacob': 80}
```

In simple words: zip() “ties together” elements from multiple sequences into grouped pairs (or triples, etc.) based on their positions.

enumerate() is a built-in Python function that gives you both the index (position) and the value when you loop over a list, tuple, string, etc.​

Simple meaning
In simple words: enumerate() adds a counter (0, 1, 2, 3, …) to each item in your sequence.​
Syntax: enumerate(iterable, start=0) – start is optional and decides from which number counting begins.​

---

```python
fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
#Here, index is the position, and fruit is the actual item.
```

With custom starting index

```python
fruits = ["apple", "banana", "cherry"]

#You can start counting from any number, not just 0 
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```
