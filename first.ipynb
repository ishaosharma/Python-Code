Named Tuple : creates a tuple subclass with named fields. and allows us access values by name instead of index - making code more readable 

```python
from collections import namedtuple

#create a namedtuple called 'Person' with fields 'name' and 'age'
Person = namedtuple('Person', ['name', 'age'])

#create an object 
p = Person('Isha', 22)

#access by name
print(p.name)
print(p.age)

#access by index (like normal tuple)
print(p[0])

print(p.name, p.age)

print(p)
```

Deque: stands for double ended queue - a list-like container that allows fast appending and popping from both ends.A queue where you can add/remove items from the left or right end very fast

```python
from collections import deque
q = deque([1,2,3])
print(q)
q.appendleft(0)
print(q)
#add element by default to right
q.append(4)
print(q)
#add element to left
q.appendleft(88)
print(q)
q.pop()   #removes from right
q.popleft()   #removes from left
print(q)  
```

```python
from collections import Counter
c = Counter(['apple', 'apple', 'banana'])
print(c['apple'])
print(c.most_common(1))
```
