```python
# 1)Write aprogram that prints all unique combinations of 1,2 and 3.
i=1
while i<=3:
    j=1
    while j<=3:
        k=1
        while k<=3:
            if i!=j and j!=k and i!=k:
                print(i,j,k)
            k+=1
        j+=1
    i+=1
```

```python
# 2)WAP that obtains decimal value of a binary numeric string.example decimal value of '1111' is 15
binary_str=input("Enter a Binary number")
decimal_value=0
power=0
# processing binary string from right to left
index=len(binary_str)-1
while index>=0:
    digit=binary_str[index]
    if digit=="1":
        decimal_value+=2**power
    elif digit!="1":
        print("Invalid binary number")
    power+=1
    index-=1
else:
    print(f"Decimal value of {binary_str} is {decimal_value}")
```

```python
# Q4) WAP to print all prime numbers between 1 and 300.
# starting from the first prime number
num=2
while num<=300:
    is_prime=True
    divisor=2

    while divisor*divisor<=num:
        if num%divisor==0:
            is_prime=False
            break
        divisor+=1

    if is_prime:
        print(num,end=" ")
    num+=1 
```

```python
# 5) WAP 2 print 24 hours of a day with suitable suffixes like AM, PM, Noon, Midnight.
hour=0
while hour<24:
    if hour==0:
        print("12 Midnight")
    elif hour==12:
        print("12 Noon")
    elif hour<12:
        print(f"{hour} AM")
    else:
        print(f"{hour-12 if hour>12 else 12}PM")
    hour+=1
```

```python
# 6) The population of a town today is 100,000.
# It has increased steadily at a rate of 10% per year for the last 10 years.
# Write a program to determine the population at the end of each year in the last decade.
population = 100000
growth_rate = 0.10
year = 1

while year <= 10:
    population = population * (1 + growth_rate)
    print(f"Population at the end of year {year}: {int(population)}")
    year += 1
```
