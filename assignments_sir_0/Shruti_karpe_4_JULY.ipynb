Q5) Program to determine the grade of a student based on their marks:

```python
marks=float(input("Enter your marks"))
# determining the grades based on the marks
if marks>90:
    # print("Grade A")
    grade="A"
elif 80<marks<89:
    # print("Grade B")
    grade="B"
elif 70<marks<79:
    # print("Grade C")
    grade="C"
elif 55<marks<69:
    # print("Grade D")
    grade="D"
else:
    # print("Grade F")
    grade="F"

#printing the result based on the grade
if grade=="F":
    print("you have failed in this course! try again")
else:
    print("Congratulatiosns! your grade is :",grade)
```
