```python
# w.a. p to find all the even and odd numbers between 0 and 40 and store them in two seperate lists
even_numbers=[]
odd_numbers=[]
for i in range(0,41):
    if i%2==0:
        even_numbers.append(i)
    else:
        odd_numbers.append(i)

print("Even numbers: ",even_numbers)
print("Odd numbers: ",odd_numbers)
```

```python
##Create an empty list. Take the input of 10 words and add them in the list.  
words=[]
for i in range(10):
    word=input(f"Enter word {i+1}: ")
    words.append(word)
print("The list of words is: ",words)
```

```python
#Use the for loop to print reverse of the list.
my_list = [1, 2, 3, 4, 5]

# Using index loop in reverse
for i in range(len(my_list) - 1, -1, -1):
    print(my_list[i])
```
