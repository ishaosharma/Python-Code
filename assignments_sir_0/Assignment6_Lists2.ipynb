1. create 2 lists of of odd numbers and even numbers with 5 elements each:
   Combine 2 lists
   add numbers 11,13,29 at the beginning of combined list
   find the number of elements in the list
   Replace the last 3 elements with 100,200,300
   Delete all the numbers in the list
   Delete the List

```python
# 1. create 2 lists of of odd numbers and even numbers with 5 elements each:
odd_numbers=[1,3,5,7,9]
even_numbers=[2,4,6,8,10]

combined=[]

# adding odd numbers individually
for i in odd_numbers:
    combined.append(i)

# adding all the even numbers at once in the combined lists
combined.extend(even_numbers)

print("Combined lists: ",combined)

#add numbers 11,13,29 at the beginning of combined list
combined.insert(0,11)
combined.insert(0,13)
combined.insert(0,29)

print("AFter adding 11,13,29 at the beginning of the lists: ",combined)

# Find number of elements in list
print("Number of elements in the lists are: ",len(combined))

# Replace last 3 elements with 100, 200, 300
combined[-3:]=[100,200,300]
print("After replacing last 3 elements: ",combined)

# deleting all the numbers in the lists
combined.clear()
print("After clearing the lists: ",combined)

# deleting the list
del combined
```

```python
# Initialize empty stack
stack = []

# Push elements onto stack
stack.append('A')
stack.append('B')
stack.append('C')

print("Stack after push operations:", stack)

# Peek at the top element (without removing)
if stack:
    print("Top element (peek):", stack[-1])
else:
    print("Stack is empty")

# Pop elements from stack (LIFO order)
if stack:
    print("Pop:", stack.pop())
if stack:
    print("Pop:", stack.pop())
if stack:
    print("Pop:", stack.pop())

print("Stack after all pop operations:", stack)

# Check if stack is empty
if not stack:
    print("Stack is empty!")
else:
    print("Stack still has elements.")
```

 WAP to generate a list of random numbers between 10 and 100.
   Delete all the elements whose value is between 20 and 50.
   Print the remaining list.


```python
# take input from the user
n=int(input("Enter the number of elements: "))
numbers=[]

print("enter numbers between 10 and 100: ")
for i in range(n):
    num=int(input(f"element {i+1}: "))
    if 10<=num<=100:
        numbers.append(num)
    else:
        print("Number out of range!")
        num = int(input(f"Element {i+1}: "))
        numbers.append(num)
print("Original Lists: ",numbers)

# removing elements from 20 and 50
filtered_lists=[x for x in numbers if not(20<=x<=50)]
#  Print remaining list
print("Filtered List (excluding values between 20 and 50):")
print(filtered_lists)
```
