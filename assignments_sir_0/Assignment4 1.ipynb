```python
# Q1)Write a program that counts the number of vowels and consonants in a given string.
str=input("Enter a string: ")
vowels="AEIOUaeuio"
vowels_cnt=0
consonant_cnt=0
for char in str:
    if char.isalpha():
        if char in vowels:
            vowels_cnt+=1
        else:
            consonant_cnt+=1
print(f"Number of vowels {vowels_cnt}")
print(f"Number of consonants {consonant_cnt}")
```

```python
# Q2)WAP Given a sentence, reverse the order of words (not characters). Example: `"Hello World"` â†’ `"World Hello"`.
sentence="Hello World"
word=sentence.split()
# reversing the list of words
reversed_word=word[::-1]

# joining the reversed words back to the sentence
reversed_sentence=" ".join(reversed_word)

print(reversed_sentence)
```

```python
# Q3) Write a  program to check if a string is a palindrome (ignoring spaces and case).

string=input("Enter a string: ")
# removing the spaces and converting it to the lowercase
# empty string to store the result
processed=""
for char in string:
    if char!=" ":
        processed+=char.lower()
        

# checking if the processed string is equal - forwards and backwards
if processed==processed[::-1]:
    print("the string is a palindrome")
else:
    print("string is not a palindrome")
```

```python
# Q4)WAP Remove duplicate characters from a string while preserving the original order.
string=input("enter a string: ")
seen=set()
result=""
for i in string:
    if i not in seen:
        seen.add(i)
        result+=i
print(result)
```

```python
# Q6)WAP Convert all lowercase letters to uppercase and vice versa in a given string
string=input("enter a string: ")
swapped=""
for char in string:
    if char.islower():
        swapped+=char.upper()
    elif char.isupper:
        swapped+=char.lower()
    else:
        # adding non aplha char directly they remain unchanged
        swapped+=char
print("String after swapping cases: ", swapped)
```

```python
# Q9)Write a program to check if two strings are anagrams of each other.
#   Note:An anagram is a word or phrase formed by rearranging the letters of another word or phrase, using all the original letters exactly once.
#    listen=silent
#    study= dusty
#    thing= night
#    all letters should be present in both words and no missing letters.
str1=input("Enter the first string: ")
str2=input("Entert the second string: ")

# remocing the spaces and converting the string to the lower case
str1_clean=str1.replace(" ","").lower()
str2_clean=str2.replace(" ","").lower()
# sorting the letters and compare
if sorted(str1_clean)==sorted(str2_clean):
    print("The strings are anagrams")
else:
    print("Strings are not anagrams")

```

```python
# Q15) Write a program to check if one string is a substring of another without using `in`.
main_str = input("Enter the main string: ")
sub_str = input("Enter the substring to check: ")

len_main = len(main_str)
len_sub = len(sub_str)

found = False

if len_sub > len_main:
    found = False
else:
    for i in range(len_main - len_sub + 1):
        match = True
        for j in range(len_sub):
            if main_str[i + j] != sub_str[j]:
                match = False
                break
        if match:
            found = True
            break

if found:
    print(f"'{sub_str}' is a substring of '{main_str}'.")
else:
    print(f"'{sub_str}' is NOT a substring of '{main_str}'.")
```
