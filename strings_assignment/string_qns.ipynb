QUE 4 : WAP to arrange string characters such that lowercase letters should come first
Given: 
str1 = pYthoNFastAPI
Expected = pthoastYNFAPI

```python
str1 = "pYthoNFastAPI"
output1 = ""
output2 = ""
for char in str1:
    if char.islower():
        output1 += char

for char in str1:
    if char.isupper():
        output2 += char

print("Output string is = ", output1+output2)


```

QUE 5: WAP to count all letters, digits and special symbols from a given string

```python
strr = "P@#yn26at^&i5ve"
count_char = 0
count_digit = 0
count_special = 0
for char in strr:
    if char.isalpha():
        count_char += 1
    elif char.isdigit():
        count_digit += 1
    else:
        count_special += 1

print("char count is = ", count_char)
print("Digit count is = ", count_digit)
print("special symbol count is = ", count_special)   
```

```python
def count_occurences(str1):
    result = {}
    for char in str1:
        if char in result:
            result[char] += 1
        else:
            result[char] = 1
    print(result)

str1 = "pythonpy"
count_occurences(str1)
```

WAP to split a given string on hyphens and display each subsstring.
text = "Violet-Indigo-Blue-Green-Yellow","Yellow","Orange","Red"
#['Violet','Indigo','Blue','Green','Yellow','Orange','Red']


```python
text = "Violet-Indigo-Blue-Green-Yellow"
t1 = "Orange"
t2 = "Red"
x = text.split("-")
# print(x) 
x.append(t1)
x.append(t2)
print(x) 
```

```python
text = "Violet-Indigo-Blue-Green-Yellow","Yellow","Orange","Red"

# Join all elements of the tuple into one string separated by commas
joined_text = ",".join(text) 
print(joined_text)

# Now split by both commas and hyphens, first replace hyphens with commas
split_colors = joined_text.replace('-', ',').split(',') 

# Remove any extra spaces and display final list
result = [color.strip() for color in split_colors] 

print(result)
```
