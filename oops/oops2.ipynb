```python
class Person:
    '''A person class'''
    name = "Amit"          #class/static variable
    age = 30

#create the object of Person class
person1 = Person()
person2 = Person()
person3 = Person()

#print Persons name
print(f"Person-1 {person1.name} age {person1.age}")
print(f"Person-2 {person2.name} age {person2.age}")
print(f"Person-3 {person1.name} age {person3.age}")

#print the id
print(id(person1.name))
print(id(person2.name))
print(id(person3.name))
```

```python
class Person:
    '''A person class'''
    name = "Amit"          #class/static variable
    age = 30

#create the object of Person class
person1 = Person()
person2 = Person()
person3 = Person()

#print Persons name
print(f"Person-1 {person1.name} age {person1.age}")
print(f"Person-2 {person2.name} age {person2.age}")
print(f"Person-3 {person1.name} age {person3.age}")

#print the id
print(id(person1.name))
print(id(person2.name))
print(id(person3.name))


#create separate variable for person1
person1.name = "Ajay"       #instance variable
print(f"Id of person-1 {id(person1.name)}")
print(f"Id of person-2 {id(person2.name)}")
print(f"Id of person-3 {id(person3.name)}")

print(f"Person-1 {person1.name} age {person1.age}")
print(f"Person-2 {person2.name} age {person2.age}")
print(f"Person-3 {person3.name} age {person3.age}")
```

```python
class Student:
    def __init__(self,name,rollno,marks): #here name,rollno,marks are not instance variable 
        #These are just argument cuz Instance variable we have to declare by using self-keyword, if we did not use self-keyword then it is not instance variable
        pass
        self.name = name        #now self.name become instance variable
        self.rollno = rollno
        self.marks = marks
        #now 3 instance variable is available
        #now we declare a method that print data related to object

    def display(self): #compulsory every instance method should contain self(instance)
        #because by using self-keyword only we can be able to access instance variables.
        print("Student Name:", self.name)
        print("Student Rollno:", self.rollno)  
        print("Student Marks:", self.marks)  
        print()

s1 = Student("abc", 101,500)
#self is implicit variable automatically add by python 
s2 = Student("xyz", 102,400)
s3 = Student("kyc", 103,700)
# 3 objects are available.
# For every object a separate copy is available in instance variable

#How to access
s1.display()  
s2.display()   
s3.display() 

#we can access by using reference name, it also works.
print(s1.name,s1.rollno,s1.marks) 
```

```python
class Employee:
    def __init__(self):
        pass
        self.amount = 200
        print("Constructor called")

    def show(self):
        self.name = "demo"

E1 = Employee()
E2 = Employee()
E3 = Employee()

E1.dep = "it"
#After the assignment, you can access E1.dep and get "it". dep property/variable is specific for only E1
# Trying to access E2.dep or E3.dep will raise an AttributeError since dep was never set for those instances.

print(E1.__dict__)  #__dict__ gives us how many instance variable are there and its corresponding values.
```
